```python
import gradio as gr
import asyncio
import nest_asyncio
from crawl4ai import AsyncWebCrawler
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_community.document_loaders import WebBaseLoader
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
from langchain_openai import ChatOpenAI
from langchain_community.vectorstores import LanceDB

nest_asyncio.apply()

# Define async function to crawl URL
async def simple_crawl(url):
    async with AsyncWebCrawler(verbose=True) as crawler:
        result = await crawler.arun(url=url)
        return result.markdown

# Define function to extract content and create retriever chain
def process_url(url):
    # Step 1: Crawl the content from the given URL
    loop = asyncio.get_event_loop()
    page_content = loop.run_until_complete(simple_crawl(url))

    # Step 2: Create document and split it
    docs = [Document(page_content=page_content, metadata={"source": url})]
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1200, chunk_overlap=300)
    splits = text_splitter.split_documents(docs)

    # Step 3: Create vectorstore and retriever
    vectorstore = LanceDB.from_documents(documents=splits, embedding=OpenAIEmbeddings())
    retriever = vectorstore.as_retriever()

    # Step 4: Set up LLM and prompt
    llm = ChatOpenAI(model="gpt-4o-mini")
    system_prompt = (
        "You are an expert assistant with access to specific information from website data. "
        "Answer user questions based on this data accurately and concisely. "
        "If the answer is not in the retrieved context, respond that the information is unavailable. "
        "Provide clear, direct answers sentences or less."
        "\n\n"
        "Context:\n\n"
        "{context}\n\n"
        "Note: Only use the information provided here."
    )
    prompt = ChatPromptTemplate.from_messages([
        ("system", system_prompt),
        ("human", "{input}"),
    ])

    # Step 5: Create chains
    question_answer_chain = create_stuff_documents_chain(llm, prompt)
    rag_chain = create_retrieval_chain(retriever, question_answer_chain)

    return rag_chain

# Define function to answer questions based on URL content
url_rag_chain = None

def handle_url_input(url):
    global url_rag_chain
    url_rag_chain = process_url(url)
    return "URL processed successfully. You can now ask questions."

def handle_question_input(question):
    if url_rag_chain is None:
        return "Please provide a URL first."
    response = url_rag_chain.invoke({"input": question})
    return response["answer"]

# Define Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("# Chatbot Q&A using Crawl4AI and LanceDB\nProvide a URL to extract content and ask questions based on it.")
    url_input = gr.Textbox(label="Enter URL", placeholder="https://example.com")
    #URL = "https://www.livemint.com/market/stock-market-news/stocks-to-buy-or-sell-for-muhurat-trading-2024-sumeet-bagadia-recommends-five-shares-to-buy-today-samvat-2081-11730449252675.html"

    url_button = gr.Button("Process URL")
    question_input = gr.Textbox(label="Ask a Question", placeholder="ask quetions based on the website or data?")
    question_button = gr.Button("Ask Question")
    output = gr.Textbox(label="Answer")

    url_button.click(fn=handle_url_input, inputs=url_input, outputs=output)
    question_button.click(fn=handle_question_input, inputs=question_input, outputs=output)

# Launch the demo
demo.launch(debug=True,share=True)
```
